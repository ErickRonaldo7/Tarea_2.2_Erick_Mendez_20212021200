```python

datos = datos[datos['housing_median_age']<40] #filtra el limite fijo de 50, casas con mayor antiguedad no se agrupen en el limite.
datos = datos[datos['median_house_value']<500000]#evita tambien que casas con mayor precio de 500k se agrupen
datos = datos[datos['median_income']<15]#evita que los salarios medios se agrupen en el limite.
datos = datos[datos['population']>7000] #filtra la poblacion de la zona. (se explicara en el informe)
```

```python
datos.hist(figsize=(15,8), bins=30, edgecolor="black")
```

```python
datos.isnull().sum() #cuenta los nulos
```

```python
datos = datos.dropna() #elimina los registros con nulos.
```

```python
dummies_ocean = pd.get_dummies(datos["ocean_proximity"], dtype=int)#convierte a dummies las categorias de ocean_proximity
```

```python
datos = pd.concat([datos, dummies_ocean], axis=1) #concatena al df original, datos.
```

```python
datos = datos.drop(["ocean_proximity"], axis=1)#elimina categorias de ocean_proximty.
```

```python
datos.drop('INLAND', axis = 1, inplace = True)
```

```python
datos["bedroom_ratio"] = datos["total_rooms"] / datos["total_bedrooms"] 

datos['habitaciones_den'] = datos['total_rooms']/ datos['households'] #habitaciones disponibles por cada area.
datos['mayor_esp_poder_adq'] = datos['median_income']/ datos['bedroom_ratio'] #habitaciones disponibles por poder adquisitivo.
datos['poder_adq'] = datos['median_income']/datos['population'] #poder adquisitivo por areas donde se encuentran las casas.
```

```python
sb.set(rc={'figure.figsize': (15,8)})
sb.heatmap(datos.corr(), annot=True, cmap="YlGnBu")
#aqui se observa que las caracteristicas anteriores tienen una correlacion imortante con la etiqueta.
```

```python
X = datos.drop(["median_house_value"], axis=1)#caracteristicas(x).
y = datos["median_house_value"]#etiqueta. (y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42) #separacion de datos prueba y entrenamiento.
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)#entrenamiento
```

```python
predicciones = modelo.predict(X_test)
```

```python
comparativa = {"Prediccion": predicciones, "Valor Real": y_test}
pd.DataFrame(comparativa)
```

```python
modelo.score(X_train, y_train) #score de datos de entrenamiento >80
```

```python
modelo.score(X_test, y_test) #score de datos de prueba.
```

```python
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error(y_test, predicciones)
```

```python
import numpy as np
rmse = np.sqrt(mse)
rmse #aplica la raÃ­z cuadrada a mse para obtener rmse
```
